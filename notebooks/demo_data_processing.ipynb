```python
### test 형식 전환 코드###

import pandas as pd

# 데이터 로드
data = pd.read_csv("2025학년도 수능 - 국어.csv")

# DataFrame으로 변환
df = pd.DataFrame(data)

# 변환 데이터를 담을 리스트
transformed_data = []

# DataFrame의 각 행을 순회하며 데이터 변환 수행
for idx, row in df.iterrows():

    # 보기 데이터를 리스트로 구성
    choices = [row["choice1"], row["choice2"], row["choice3"], row["choice4"], row["choice5"]]
    
    # 문제 데이터 구성
    if "question_plus" in row and pd.notna(row["question_plus"]):
        problem = {
            "question": row["question"],
            "choices": choices,
            "answer": '',
            "question_plus": row["question_plus"]
        }
    else:
        problem = {
            "question": row["question"],
            "choices": choices,
            "answer": ''
        }
    
    # 최종 데이터 구성
    transformed_data.append({
        "id": row["id"],
        "paragraph": row["paragraph"],
        "problems": problem,
        "question_plus": row["question_plus"]
    })

# 변환된 데이터프레임 생성
transformed_df = pd.DataFrame(transformed_data)

# 파일 저장
output_file = "2025ksat_korean.csv"
transformed_df.to_csv(output_file, index=False)
```

```python
### output 형식 전환 코드 ###

import pandas as pd

# 데이터 로드
data = pd.read_csv('2025학년도 수능 - 국어.csv')

# DataFrame으로 변환
df = pd.DataFrame(data)

# 변환 데이터를 담을 리스트
transformed_data = []

# DataFrame의 각 행을 순회하며 데이터 변환 수행
for idx, row in df.iterrows():

    # 최종 데이터 구성
    transformed_data.append({
        "id": row["id"],
        "answer": row["answer"]
    })

# 변환된 데이터프레임 생성
transformed_df = pd.DataFrame(transformed_data)

# 파일 저장
output_file = "2025ksat_korean_answer.csv"
transformed_df.to_csv(output_file, index=False)
```

```python
### 전체 데이터 병합 코드 ###

import pandas as pd

# CSV 파일 경로 지정
file1 = '2025학년도 수능 - 국어 (1).csv'
file2 = '2025학년도 수능 - 동아시아사 (1).csv'
file3 = '2025학년도 수능 - 사회문화 (1).csv'
file4 = '2025학년도 수능 - 생활과 윤리.csv'
file5 = '2025학년도 수능 - 세계사 (1).csv'
file6 = '2025학년도 수능 - 윤리와 사상 (1).csv'
file7 = '2025학년도 수능 - 정치와법 (1).csv'
file8 = '2025학년도 수능 - 한국사 (1).csv'

# 데이터프레임 전환
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)
df5 = pd.read_csv(file5)
df6 = pd.read_csv(file6)
df7 = pd.read_csv(file7)
df8 = pd.read_csv(file8)

# 데이터프레임을 하나로 병합
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)

# 파일 저장
merged_df.to_csv('train_ksat.csv', index=False)
```
